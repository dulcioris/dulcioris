```python
import random
import networkx as nx
import matplotlib.pyplot as plt

def sim_red_metabolica(n):
    metabolitos = ["A", "B", "C", "D"]  # Ejemplo de metabolitos
    red_metabolica = []

    for i in range(n):
        metabolito = random.choice(metabolitos)
        red_metabolica.append(metabolito)

    return red_metabolica

def create_red_metabolica_graph(red_metabolica):
    G = nx.Graph()

    # Agregar nodos para cada metabolito
    for i, metabolito in enumerate(red_metabolica):
        G.add_node(i, metabolito=metabolito)

    # Conectar nodos adyacentes para representar la red metab√≥lica
    for i in range(len(red_metabolica) - 1):
        G.add_edge(i, i+1)

    return G

def plot_red_metabolica_graph(G):
    pos = {node: (node, 0) for node in G.nodes()}
    labels = nx.get_node_attributes(G, 'metabolito')

    nx.draw(G, pos, with_labels=True, labels=labels, font_color='white', font_size=8, node_color='lightgreen', node_size=700)
    plt.show()

# Ejemplo de uso
red_metabolica_ejemplo = sim_red_metabolica(10)
graph_metabolico = create_red_metabolica_graph(red_metabolica_ejemplo)
plot_red_metabolica_graph(graph_metabolico)
```

```python
# pip install networkx
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random

seed = 0
random.seed(seed)
np.random.seed(seed)
G = nx.Graph()
G.graph["Name"] = "Red Metab√≥lica"

G.add_nodes_from(["Acetil-CoA", "Oxalacetato", "Citrato", "Cis-Aconitato", "Isocitrato",
                 "Alfa-cetoglutarato", "Succinil-CoA", "Succinato", "Fumarato", "Malato"])

G.add_edges_from([("Fumarato", "Malato"), ("Malato", "Oxalacetato"), ("Oxalacetato", "Acetil-CoA"), ("Acetil-CoA", "Citrato"),
                 ("Citrato", "Cis-Aconitato"), ("Cis-Aconitato", "Alfa-cetoglutarato"), ("Alfa-cetoglutarato", "Succinil-CoA"),
                 ("Succinil-CoA", "Succinato"), ("Succinato", "Fumarato"),
                 ("Cis-Aconitato", "Isocitrato")])

pos = {
    "Acetil-CoA": (4, 1),
    "Oxalacetato": (3, 3),
    "Citrato": (5, 4),
    "Cis-Aconitato": (7, 8),
    "Isocitrato": (10, 7),
  #  "Oxalosuccinato"#:# (2, 4),
    "Alfa-cetoglutarato": (8, 6),
    "Succinil-CoA": (6, 1),
    "Succinato": (8, 3),
    "Fumarato": (10, 4),
    "Malato": (3, 7)
}

nx.draw(G, with_labels=True, node_color="lightblue",
        node_size=3000, font_color="black", font_size=12,
        font_family="Times New Roman", font_weight="bold", width=2, pos=pos)
plt.margins(0.2)
plt.title("Ciclo de Krebs", fontsize=16, fontweight="bold")
plt.figure(figsize=(12, 8))  # Ajusta el tama√±o de la figura seg√∫n tus preferencias
plt.subplots_adjust(wspace=0.4)  # Ajusta el espacio horizontal entre subgr√°ficos
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random

seed = 0
random.seed(seed)
np.random.seed(seed)

G = nx.Graph()
G.graph["Name"] = "Ciclo de la Gluc√≥lisis"

# Nombres de los compuestos en el ciclo de la gluc√≥lisis
compuestos_glucolisis = ["Glucosa", "Glucosa-6-Fosfato", "Fructosa-6-Fosfato", "Fructosa-1,6-Bifosfato",
                         "Gliceraldehido-3-Fosfato", "Glicerato-1,3-Bifosfato", "Glicerato-3-Fosfato",
                         "Fosfoenolpiruvato", "Piruvato"]

G.add_nodes_from(compuestos_glucolisis)

# Definici√≥n de los bordes del ciclo de la gluc√≥lisis
bordes_glucolisis = [
    ("Glucosa", "Glucosa-6-Fosfato"),
    ("Glucosa-6-Fosfato", "Fructosa-6-Fosfato"),
    ("Fructosa-6-Fosfato", "Fructosa-1,6-Bifosfato"),
    ("Fructosa-1,6-Bifosfato", "Gliceraldehido-3-Fosfato"),
    ("Gliceraldehido-3-Fosfato", "Glicerato-1,3-Bifosfato"),
    ("Glicerato-1,3-Bifosfato", "Glicerato-3-Fosfato"),
    ("Glicerato-3-Fosfato", "Fosfoenolpiruvato"),
    ("Fosfoenolpiruvato", "Piruvato")
]

G.add_edges_from(bordes_glucolisis)

# Posiciones de los nodos en el gr√°fico
pos_glucolisis = {
    "Glucosa": (1, 5),
    "Glucosa-6-Fosfato": (2, 4),
    "Fructosa-6-Fosfato": (3, 3),
    "Fructosa-1,6-Bifosfato": (4, 2),
    "Gliceraldehido-3-Fosfato": (5, 8),
    "Glicerato-1,3-Bifosfato": (6, 5),
    "Glicerato-3-Fosfato": (7, 6),
    "Fosfoenolpiruvato": (8, 7),
    "Piruvato": (9, 8)
}

nx.draw(G, with_labels=True, node_color="lightgreen",
        node_size=3000, font_color="black", font_size=12,
        font_family="Times New Roman", font_weight="bold", width=2, pos=pos_glucolisis)
plt.title("Ciclo de la Gluc√≥lisis",fontsize=16, fontweight="bold")
plt.margins(0.2)
plt.show()
```
