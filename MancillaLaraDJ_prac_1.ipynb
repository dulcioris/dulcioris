# 1. Resuelva los siguientes problemas de valor inicial de forma analítica.

**a)** **$y’ = y^2 t^2, y(0)=1$**

$\frac{dy}{dt} = t^2 \cdot y^2$, $( y(0) = 1 )$
Primero, separaremos las variables $y$ y $t$:

$\frac{dy}{y^2} = t^2 \, dt.$

Integramos ambos lados de la ecuación:

$\int \frac{dy}{y^2} = \int t^2 \, dt.$

La integral del lado izquierdo es:

$\int y^{-2} \, dy = -y^{-1} = -\frac{1}{y}.$

La integral del lado derecho es:

$\int t^2 \, dt = \frac{t^3}{3}.$

La ecuación resultante sería:

$-\frac{1}{y} = \frac{t^3}{3} + C,$

donde $C$ es la constante de integración.

Ahora, despejamos $y$ de la ecuación:

$\frac{1}{y} = -\frac{t^3}{3} - C.$

Invirtiendo ambos lados:

$y(t) = \frac{1}{- \frac{t^3}{3} - C}.$

Usamos la condición inicial $y(0) = 1$ para encontrar $C$:

$1 = \frac{1}{- \frac{0^3}{3} - C} = \frac{1}{-C}.$

Entonces, $C = -1$.

$y(t) = \frac{1}{1 - \frac{t^3}{3}}.$

La solución es:

$y(t) = \frac{1}{1 - \frac{t^3}{3}}.$

**b)** **$y’= 5t + y$**
$\frac{dy}{dt} = 5t + y$, $(y(0) = 1)$
La ecuación diferencial se puede reescribir como:

$\frac{dy}{dt} - y = 5t.$

Aquí, identificamos \( P(t) = -1 \) y \( Q(t) = 5t \).

$\mu(t) = e^{\int P(t) \, dt} = e^{\int -1 \, dt} = e^{-t}.$

Multiplicamos toda la ecuación diferencial por  $\mu(t) = e^{-t} $:

$e^{-t} \frac{dy}{dt} - e^{-t} y = 5t e^{-t}.$

El lado izquierdo de la ecuación es ahora la derivada del producto $ y(t) \cdot e^{-t} $:

$\frac{d}{dt} \left( y(t) \cdot e^{-t} \right) = 5t e^{-t}.$

Integramos ambos lados con respecto a \(t\):

$\int \frac{d}{dt} \left( y(t) \cdot e^{-t} \right) \, dt = \int 5t e^{-t} \, dt.$

El lado izquierdo se simplifica a:

$y(t) \cdot e^{-t}.$

Para integrar el lado derecho, utilizamos la integración por partes:

Sea $ u = 5t $ y $ dv = e^{-t} dt $.

Entonces, $ du = 5 dt $ y $ v = -e^{-t} $

Ahora, aplicamos la integración por partes:

$\int 5t e^{-t} \, dt = -5t e^{-t} + 5 \int e^{-t} \, dt = -5t e^{-t} - 5 e^{-t} = -5e^{-t}(t + 1) + C$

Por lo tanto, la ecuación se convierte en:

$y(t) \cdot e^{-t} = -5e^{-t}(t + 1) + C.$

Multiplicamos ambos lados por $e^{t}$ para despejar $y(t)$:

$y(t) = -5(t + 1) + C e^{t}.$

Aplicamos la condición inicial $y(0) = 1$:

$1 = -5(0 + 1) + C e^{0} = -5 + C.$

Entonces, $C = 6$

$y(t) = -5(t + 1) + 6e^{t}.$

La solución del PVI es:

$y(t) = 6e^t - 5(t + 1).$

$y_{n+1} = y_n + h \cdot (t_n^2 \cdot y_n^2)$

$y_{n+1} = y_n + h \cdot \left((nh)^2 \cdot y_n^2\right).$

$y_{n+1} = y_n + h \cdot (5nh + y_n)$

##**Método de Euler.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(t, y) para cada problema

# Problema 1: dy/dt = t^2 * y^2
def problem1_f(t, y):
    return t**2 * y**2

# Problema 2: dy/dt = 5t + y
def problem2_f(t, y):
    return 5*t + y

# Método de Euler
def euler(f, t0, y0, h, tf):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0

    for i in range(n):
        y[i+1]=y[i]+h*f(t[i],y[i])

    return t, y

# Parámetros del problema
t0 = 0
y0 = 1
tf = 10

# Valores de h para probar
h_values = [0.1, 0.01, 0.001, 0.0001]

# Graficar los resultados para cada h

plt.figure(figsize=(12, 8))

for h in h_values:
    t, y_problem1 = euler(problem1_f, t0, y0, h, tf)
    t, y_problem2 = euler(problem2_f, t0, y0, h, tf)

    plt.subplot(2, 1, 1)
    plt.plot(t, y_problem1, label=f'h = {h}')
    plt.title('Aproximación por método de Euler - Problema 1: dy/dt = t^2 * y^2')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(t, y_problem2, label=f'h = {h}')
    plt.title('Aproximación por método de Euler - Problema 2: dy/dt = 5t + y')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(t, y) para el Problema 1
def problem1_f(t, y):
    return t**2 * y**2

# Solución exacta para el Problema 1
def exact_solution_problem1(t):
    # La solución exacta de la ecuación diferencial es:
    # y(t) = 1 / (1 - t^3 / 3) para |t| < (3^(1/3))
    # La solución es válida hasta t = 3^(1/3), momento en que la función se vuelve indefinida.
    return 1 / (1 - (t**3) / 3)

# Método de Euler
def euler(f, t0, y0, h, tf):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0

    for i in range(n):
        y[i + 1] = y[i] + h * f(t[i], y[i])

    return t, y

# Parámetros del problema
t0 = 0
y0 = 1
tf = 10

# Valores de h para probar
h_values = [0.1, 0.01, 0.001, 0.0001]

# Graficar los errores de truncamiento global para cada h

plt.figure(figsize=(12, 8))

for h in h_values:
    t, y_aprox_problem1 = euler(problem1_f, t0, y0, h, tf)
    y_exacta_problem1 = exact_solution_problem1(t)

    # Debido a la naturaleza del problema, la solución exacta podría divergir para t > 3^(1/3)
    # Lo que provoca que la función se vuelva indefinida. Para evitar errores numéricos
    # recortamos t en 3^(1/3)
    valid_indices = t <= 3**(1/3)
    t_valid = t[valid_indices]
    y_aprox_valid = y_aprox_problem1[valid_indices]
    y_exacta_valid = y_exacta_problem1[valid_indices]

    error_truncamiento_problem1 = np.abs(y_exacta_valid - y_aprox_valid)

    plt.plot(t_valid, error_truncamiento_problem1, label=f'Problema 1: h={h}')

plt.title('Error global de truncamiento - Problema 1')
plt.xlabel('t')
plt.ylabel('Error')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(t, y) para el Problema 1
def problem1_f(t, y):
    return t**2 * y**2

# Solución exacta para el Problema 1
def exact_solution_problem1(t):
    # Solución exacta teórica
    return 1 / (1 - (t**3) / 3)

# Método de Euler
def euler(f, t0, y0, h, tf):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0

    for i in range(n):
        y[i + 1] = y[i] + h * f(t[i], y[i])
        # Manejo de la singularidad: si estamos cerca de la singularidad, detener el cálculo
        if abs(1 - (t[i]**3) / 3) < 1e-6:
            t = t[:i+2]
            y = y[:i+2]
            break

    return t, y

# Parámetros del problema
t0 = 0
y0 = 1
tf = 10

# Valores de h para probar
h_values = [0.1, 0.01, 0.001, 0.0001, 0.00001]

# Graficar los errores de truncamiento global para cada h

plt.figure(figsize=(12, 8))

for h in h_values:
    t, y_aprox_problem1 = euler(problem1_f, t0, y0, h, tf)
    y_exacta_problem1 = exact_solution_problem1(t)

    # Cálculo del error
    error_truncamiento_problem1 = np.abs(y_exacta_problem1 - y_aprox_problem1)

    plt.plot(t, error_truncamiento_problem1, label=f'Problema 1: h={h}')

plt.title('Error global de truncamiento - Problema 1')
plt.xlabel('t')
plt.ylabel('Error')
plt.legend()
plt.grid(True)
plt.show()
```

Lo que cambié en el código fue:
1. **Método de Euler Mejorado**: En el bucle del método de Euler, intenté añadir una condición que detuviera la ejecución si estamos cerca de la singularidad en $t = \sqrt[3]{3}$.

2. **Valores Pequeños de h**: Introduje $h$ cada vez más pequeños, incluso hasta $h = 0.00001$. Esto permitirá observar cómo se comporta la solución numérica a medida que nos acercamos a la singularidad.

3. **Manejo de la Singularidad**: Al detener el cálculo antes de que la función exacta se vuelva indefinida, el código evita que se generen errores numéricos significativos.

##**b)Comparación entre la solución exacta y las aproximadas.**



```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las funciones f(t, y) para los Problemas 1 y 2
def problem1_f(t, y):
    return t**2 * y**2

def problem2_f(t, y):
    return 5*t + y

# Soluciones exactas para los Problemas 1 y 2
def exact_solution_problem1(t):
    return 1 / (1 - (t**3) / 3)

def exact_solution_problem2(t):
    return 25*np.exp(t) - 5*t - 24

# Método de Euler
def euler(f, t0, y0, h, tf):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0

    for i in range(n):
        y[i + 1] = y[i] + h * f(t[i], y[i])

    return t, y

# Parámetros del problema
x0 = 0
y0 = 1
xf = 10

# Valores de h para probar
h_values = [0.1, 0.05, 0.01]

# Gráfico para el Problema 1
x = np.linspace(x0, xf, 1000)
y_exacta1 = exact_solution_problem1(x)

plt.figure(figsize=(12, 8))
plt.plot(x, y_exacta1, 'k-', label='Solución Exacta Problema 1')

for h in h_values:
    x_approx, y_approx1 = euler(problem1_f, x0, y0, h, xf)
    plt.plot(x_approx, y_approx1, label=f'Solución Aproximada (h = {h})')

plt.title("Comparación entre la Solución Exacta y las Soluciones Aproximadas - Problema 1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Gráfico para el Problema 2
y_exacta2 = exact_solution_problem2(x)

plt.figure(figsize=(12, 8))
plt.plot(x, y_exacta2, 'k-', label='Solución Exacta Problema 2')

for h in h_values:
    x_approx, y_approx2 = euler(problem2_f, x0, y0, h, xf)
    plt.plot(x_approx, y_approx2, label=f'Solución Aproximada (h = {h})')

plt.title("Comparación entre la Solución Exacta y las Soluciones Aproximadas - Problema 2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las funciones f(t, y) para los Problemas 1 y 2
def problem1_f(t, y):
    return t**2 * y**2

def problem2_f(t, y):
    return 5*t + y

# Soluciones exactas para los Problemas 1 y 2
def exact_solution_problem1(t):
    return 1 / (1 - (t**3) / 3)

def exact_solution_problem2(t):
    return 25*np.exp(t) - 5*t - 24

# Método de Euler
def euler(f, t0, y0, h, tf):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0

    for i in range(n):
        y[i + 1] = y[i] + h * f(t[i], y[i])

    return t, y

# Parámetros del problema
x0 = 0
y0 = 1
xf = 10

# Valores de h para probar
h_values = [0.1, 0.01, 0.001, 0.0001]

# Gráfico de Error Absoluto para el Problema 1
plt.figure(figsize=(12, 8))
x = np.linspace(x0, xf, 1000)
y_exacta1 = exact_solution_problem1(x)

for h in h_values:
    x_approx, y_approx1 = euler(problem1_f, x0, y0, h, xf)
    y_exacta_approx = np.interp(x_approx, x, y_exacta1)
    Error_abs1 = np.abs(y_exacta_approx - y_approx1)
    plt.plot(x_approx, Error_abs1, label=f'h = {h}')

plt.title("Error Absoluto para Diferentes Valores de h - Problema 1")
plt.xlabel("x")
plt.ylabel("Error Absoluto")
plt.legend()
plt.grid(True)
plt.show()

# Gráfico de Error Absoluto para el Problema 2
plt.figure(figsize=(12, 8))
y_exacta2 = exact_solution_problem2(x)

for h in h_values:
    x_approx, y_approx2 = euler(problem2_f, x0, y0, h, xf)
    y_exacta_approx = np.interp(x_approx, x, y_exacta2)
    Error_abs2 = np.abs(y_exacta_approx - y_approx2)
    plt.plot(x_approx, Error_abs2, label=f'h = {h}')

plt.title("Error Absoluto para Diferentes Valores de h - Problema 2")
plt.xlabel("x")
plt.ylabel("Error Absoluto")
plt.legend()
plt.grid(True)
plt.show()
```
