```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo en forma de cuadrícula 5x5
G = nx.grid_2d_graph(5, 5, periodic=False)

# Asignar automáticamente el valor del peso de cada arista como 5
for edge in G.edges:
    G.edges[edge]["weight"] = 5

# Generar automáticamente las letras A a Z para asignar a los nodos
alphabet = list(map(chr, range(65, 91)))
node_labels = {node: label for node, label in zip(G.nodes, alphabet)}

# Dibujar el grafo como una cuadrícula con etiquetas automáticas
pos = {(i, j): (j, -i) for i, j in G.nodes}
nx.draw(G, pos, with_labels=True, labels=node_labels, node_color="red",
        node_size=3000, font_color="white", font_size=10,
        font_family="serif", font_weight="bold", width=5)

# Mostrar el grafo
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo en forma de cuadrícula 5x5 más un elemento
G = nx.grid_2d_graph(5, 5, periodic=False)

# Asignar automáticamente el valor del peso de cada arista como 5
for edge in G.edges:
    G.edges[edge]["weight"] = 5

# Generar automáticamente las letras A a Z (agregando Z al final)
alphabet = list(map(chr, range(65, 91))) + ['Z']
node_labels = {node: label for node, label in zip(G.nodes, alphabet)}

# Dibujar el grafo como una cuadrícula con etiquetas automáticas
pos = {(i, j): (j, -i) for i, j in G.nodes}
nx.draw(G, pos, with_labels=True, labels=node_labels, node_color="red",
        node_size=3000, font_color="white", font_size=10,
        font_family="serif", font_weight="bold", width=5)

# Mostrar el grafo
plt.show()
```
